# Introduction to Python

There are four collection data types in the Python programming language:
1. List : ordered and changeable. Allows duplicate members.
2. Tuple : ordered and unchangeable. Allows duplicate members.
3. Set : unordered and changeable. No duplicate members.
4. Dictionary: unordered, changeable. No duplicate members.

### List 

List is a collection which is ordered and changeable. Allows duplicate members.

```python
#creating a list

thislist = ["apple", "banana", "cherry"]
print(thislist)
```

```python
#Access items 
thislist = ["apple", "banana", "cherry"]
print(thislist[2])
```

```python
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

```python
# slicing, a:b, index a is included, index b is not included

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])     
```

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])

# As early as possible it can start
```

```python
#Loop Through a List

thislist = ["apple", "banana", "cherry"]
for x in thislist:
    print(x)
```

```python
#length of list

thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

```python
#add item in the list using append() method.

thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
#add item in the list using insert() method at specified location.


thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)

# Item at passed index moves to i + 1, where i is the passed index
```

```python
#remove item from list.

thislist = ["apple", "banana", "cherry", "banana"]
thislist.remove("banana")
print(thislist)
print(thislist.count("banana"))

# first entry equal to passed, gets removed
```

```python
#remove item from specified location in list.

thislist = ["apple", "banana", "cherry"]
#thislist.pop()
thislist.pop(1)
print(thislist)
```

```python
#clear method make list empty.

thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)
```

```python
#copy a list

thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
```

```python
#join 2 list or concatenate two lists

list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
```

```python
#list() constructor to make a new list. Here it creates a list from a tuple.

thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)

# here a tuple is getting converted into tuple
```

Other built-in methods..

* append() Adds an element at the end of the list
* count()	 Returns the number of elements with the specified value
* index()	 Returns the index of the first element with the specified value
* insert() Adds an element at the specified position
* pop()	 Removes the element at the specified position
* remove() Removes the item with the specified value
* reverse() Reverses the order of the list
* sort()    Sort the elements

### Tuple
A tuple is a collection which is ordered and unchangeable. In Python tuples are written with round brackets.

```python
#create tuple

thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

```python
#access tuple items
# as it is ordered, indexing works just fine

thistuple = ("apple", "banana", "cherry")
print(thistuple[1])
```

```python
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])

# Slicing works find
```

```python
#Loop throught a Tuple

thistuple = ("apple", "banana", "cherry")
for x in thistuple:
    print(x)
```

```python
#Tuple length

thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
```

```python
#Add item in tuple.
'''
It will give you error. you can't add items in tuple because it is unchangeable.
'''
thistuple = ("apple", "banana", "cherry")
thistuple[2] = "orange" # This will raise an error
print(thistuple)
```

```python
#join two tuple

tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)

# Concatenation works fine
```

### Set
A set is a collection which is unordered and changeable. In Python sets are written with curly brackets. Set doesn't allow duplicates.

```python
#creating set

thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
#Access item in a set.
'''
You can't access item in a set by referring to an index, 
since sets are unordered the items has no index.
'''
thisset = {"apple", "banana", "cherry"}
#print(thisset[0])
for x in thisset:
    print(x)
```

```python
#Chagne item add(), update() method.

thisset = {"apple", "banana", "cherry"}

#thisset.add("orange") # Works for adding single element
thisset.update(["orange", "mango", "grapes", "orange"])
print(thisset)
```

```python
#remove item using remove() or discard() method.
'''

If the item to remove does not exist, remove() will raise an error.
If the item to remove does not exist, discard() will NOT raise an error.

'''

thisset = {"apple", "cherry"}

#thisset.remove("banana")
thisset.discard("banana")

print(thisset)
```

```python
#Join two sets using union() or update()
'''
union() method that returns a new set containing all items from both sets.
update() method that inserts all the items from one set into another
'''

set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)

#set1.update(set2)
#print(set1)
print(set3)
```

# Dictionary

```python

# Create and print a dictionary:

dict1 = {
  "1": "Ram",
  "Gender": "Male",
  "year": 1998
}
print(dict1)
```

```python
dict2 = dict(name="Ford", Gender="Male", year=1964)
dict2
```

```python
#Accessing Items, Dictionary is ordered

x = dict1["1"]
x = dict1.get("year")
print(x)
```

```python
# Change Values, Assignment operatoin works just fine

dict1["year"] = 2020
dict1
```

```python
# Loop Through a Dictionary

# Print all key names in the dictionary:

for x in dict1:
  print(x)
```

```python
# Print all values in the dictionary, one by one:

for x in dict1:
  print(dict1[x])
```

```python
# print both key and value

for x, y in dict1.items():
  print(x, y)

dict1.items()

# dict1.items() is a list storing key value pairs of whole Dictionary
```

```python
#Adding Items

dict1["Age"] = 25

print(dict1)
```

```python
#Removing Items

dict1.pop("Age")

dict1.popitem()

#del dict1["Age"]

dict1
```
