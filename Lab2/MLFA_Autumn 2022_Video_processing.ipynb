```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Opening a video file, extracting and operating on its frames.

```python
# Create VideoCapture object to read the video frame by frame
cap = cv2.VideoCapture('ballet_dancer.mp4')
if (cap.isOpened() == False):
    print("Error opening video")
else:
    fps = cap.get(5)  #get frame rate
    print('Frames per second : ', fps,'FPS')
    frame_count = cap.get(7) #get frame count
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    frame_size = (frame_width,frame_height)
    print('Frame count : ', frame_count) 
```

```python
frame_size
```

```python
i=0
while(cap.isOpened() and i<5):
    ret, frame = cap.read() #get frame along with a boolean value
    if ret == True:
        cv2.imwrite('frame_'+str(i+1)+'.png',frame)
        canny = cv2.Canny(frame,10,100)#running edge detector with min and max thresholds
        cv2.imwrite('edg_frame_'+str(i+1)+'.png',canny)#saving the edge-detected image
        
            #other transformations/operations on individual frames
            
        i+=1
```

Writing a video file

```python
out = cv2.VideoWriter('sample_vid.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 24, frame_size)
```

```python
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        frame = cv2.flip(frame,0)
        out.write(frame)
    else:
        print('Stream disconnected')
        break
```

```python
# Release the video capture object
cap.release()
out.release()
cv2.destroyAllWindows()
```
