## Gradient Descent using Single Independent Variable

```python
import numpy as np
import matplotlib.pyplot as plt

def func_x(x):
    return (x+5) ** 2

x = np.linspace(-10, 0, 30)

z = func_x(x)

plt.plot(x, z)
```

```python
import numpy as np
import matplotlib.pyplot as plt

cur_x = 3
learning_rate = 0.01
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0

f = lambda x: (x+5)**2
df = lambda x: 2*(x+5)

x_values_gradient_descent = []

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - learning_rate * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)
    iters = iters + 1
    if (iters % 50) == 0:
      print("Iteration",iters,"\nX value is",cur_x)

    x_values_gradient_descent.append(cur_x)
    
print("The local minimum occurs at", cur_x)

print(x_values_gradient_descent)

x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = f(x_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
```

## Gradient Descent using Multiple Independent Variables

```python
import numpy as np
import matplotlib

def func_x_y(x, y):
    return 1 + (0.1 * (y**2 + (2*(x**2))))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = func_x_y(X, Y)

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
ax.set_zlim3d(0, 15)
```

```python
import numpy as np
import matplotlib.pyplot as plt

cur_x = 3
cur_y = 40
learning_rate = 0.1
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0

f = lambda x,y: 1 + (0.1 * (y**2 + (2*(x**2))))
df_dx = lambda x,y: 0.1 * 2 * 2 * x
df_dy = lambda x,y: 0.1 * 2 * y

x_values_gradient_descent = []
y_values_gradient_descent = []

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x
    prev_y = cur_y
    cur_x = cur_x - learning_rate * df_dx(prev_x, prev_y)
    cur_y = cur_y - learning_rate * df_dy(prev_x, prev_y)
    previous_step_size = (((cur_x - prev_x) ** 2) + ((cur_y - prev_y) ** 2)) ** 0.5
    iters = iters + 1
    if (iters % 50) == 0:
      print("Iteration",iters,"\nX value is",cur_x,"\nY value is",cur_y)

    x_values_gradient_descent.append(cur_x)
    y_values_gradient_descent.append(cur_y)
    
print("The local minimum occurs at", cur_x, cur_y)

print(x_values_gradient_descent)
print(y_values_gradient_descent)

x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = np.array(y_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
```

## Gradient Descent (Momentum) using Multiple Independent Variables

```python
import numpy as np
import matplotlib

def func_x_y(x, y):
    return 1 + (0.1 * (y**2 + (2*(x**2))))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = func_x_y(X, Y)

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
ax.set_zlim3d(0, 15)
```

```python
import numpy as np
import matplotlib.pyplot as plt

cur_x = 3
cur_y = 40
learning_rate = 0.1
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0

gamma = 0.9

f = lambda x,y: 1 + (0.1 * (y**2 + (2*(x**2))))
df_dx = lambda x,y: 0.1 * 2 * 2 * x
df_dy = lambda x,y: 0.1 * 2 * y

x_values_gradient_descent = []
y_values_gradient_descent = []

prev_del_x = 0.0
prev_del_y = 0.0

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x
    prev_y = cur_y
    cur_x = cur_x - ((learning_rate * df_dx(prev_x, prev_y)) + (gamma * prev_del_x))
    cur_y = cur_y - ((learning_rate * df_dy(prev_x, prev_y)) + (gamma * prev_del_y))

    prev_del_x = learning_rate * df_dx(prev_x, prev_y)
    prev_del_y = learning_rate * df_dy(prev_x, prev_y)

    previous_step_size = (((cur_x - prev_x) ** 2) + ((cur_y - prev_y) ** 2)) ** 0.5
    iters = iters + 1
    if (iters % 50) == 0:
      print("Iteration",iters,"\nX value is",cur_x,"\nY value is",cur_y)

    x_values_gradient_descent.append(cur_x)
    y_values_gradient_descent.append(cur_y)
    
print("The local minimum occurs at", cur_x, cur_y)

print(x_values_gradient_descent)
print(y_values_gradient_descent)

x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = np.array(y_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
```

## Gradient Descent: ADAM (Adaptive Moment Estimation) using Multiple Independent Variables

```python
import numpy as np
import matplotlib

def func_x_y(x, y):
    return 1 + (0.1 * (y**2 + (2*(x**2))))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = func_x_y(X, Y)

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
ax.set_zlim3d(0, 15)
```

```python
import numpy as np
import matplotlib.pyplot as plt

cur_x = 3
cur_y = 40
learning_rate = 0.1
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0

beta_1 = 0.9
beta_2 = 0.999

f = lambda x,y: 1 + (0.1 * (y**2 + (2*(x**2))))
df_dx = lambda x,y: 0.1 * 2 * 2 * x
df_dy = lambda x,y: 0.1 * 2 * y

x_values_gradient_descent = []
y_values_gradient_descent = []

prev_m_t_x = 0.0
prev_m_t_y = 0.0
prev_v_t_x = 0.0
prev_v_t_y = 0.0

epsilon = 1e-8

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x
    prev_y = cur_y

    gradient_x = df_dx(prev_x, prev_y)
    gradient_y = df_dy(prev_x, prev_y)

    m_t_x = ((1.0 - beta_1) * gradient_x) + (beta_1 * prev_m_t_x)
    m_t_y = ((1.0 - beta_1) * gradient_y) + (beta_1 * prev_m_t_y)
    v_t_x = ((1.0 - beta_2) * (gradient_x ** 2)) + (beta_2 * prev_v_t_x)
    v_t_y = ((1.0 - beta_2) * (gradient_y ** 2)) + (beta_2 * prev_v_t_y)

    prev_m_t_x = m_t_x
    prev_m_t_y = m_t_y
    prev_v_t_x = v_t_x
    prev_v_t_y = v_t_y

    m_cap_t_x = m_t_x / (1 - (beta_1 ** (iters + 1)))
    m_cap_t_y = m_t_y / (1 - (beta_1 ** (iters + 1)))
    v_cap_t_x = v_t_x / (1 - (beta_2 ** (iters + 1)))
    v_cap_t_y = v_t_y / (1 - (beta_2 ** (iters + 1)))

    cur_x = cur_x - ((learning_rate * m_cap_t_x) / ((v_cap_t_x ** 0.5) + epsilon))
    cur_y = cur_y - ((learning_rate * m_cap_t_y) / ((v_cap_t_y ** 0.5) + epsilon))

    previous_step_size = (((cur_x - prev_x) ** 2) + ((cur_y - prev_y) ** 2)) ** 0.5
    iters = iters + 1
    if (iters % 50) == 0:
      print("Iteration",iters,"\nX value is",cur_x,"\nY value is",cur_y)

    x_values_gradient_descent.append(cur_x)
    y_values_gradient_descent.append(cur_y)
    
print("The local minimum occurs at", cur_x, cur_y)

print(x_values_gradient_descent)
print(y_values_gradient_descent)

x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = np.array(y_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
```
