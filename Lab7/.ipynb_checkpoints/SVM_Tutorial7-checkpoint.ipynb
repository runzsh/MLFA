**Support Vector Machine (SVM)**

It is a discriminative classifier capable of performing classification, regression and even outlier detection. 

The linear SVM classifier separates the two classes by drawing line between them also called hyperplane or decision boundary. It can define both a linear or a non-linear decision boundary by using kernel functions. All the data points that fall on one side of the line will be labeled as one class and all the points that fall on the other side will be labeled as the second.

It draws a margin around the decision boundary to separate the vectors from the decision boundary, as much as possible which gives us more confidence in our predictions. Because the vectors are at least the length of the margin away from the decision boundary, thereâ€™s less ambiguity during classification. 

The position of the margin is defined using the vectors that are closest to the decision boundary. Support vectors are the vectors that lie on top of the margin.

```python
#importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs # (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html)
from sklearn.svm import SVC #Support Vector Classifier (https://scikit-learn.org/stable/modules/svm.html)
```

```python
#generate sample data

X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.6)
print(X.shape)
print(y)
plt.scatter(X[:,0], X[:,1], c=y);
```

```python
#model fitting

model = SVC(kernel='linear', C=1)
model.fit(X,y)
#C is Regularization parameter. The strength of the regularization is inversely proportional to C. 
#Must be strictly positive. Deafault value of C is 1
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca() # gets the current axes; creates an empty plot
    xlim = ax.get_xlim() 
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30) # Returns 30 evenly spaced samples, calculated over the interval [xlim[0], xlim[1]].
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x) # creates a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing
    xy = np.vstack([X.ravel(), Y.ravel()]).T # Stacking flattened arrays in sequence vertically
    P = model.decision_function(xy).reshape(X.shape) # Returns the decision function of the sample for each class in the model
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, levels=[-1, 0, 1], linestyles=['--', '-', '--']) # levels determines the number and positions of the contour lines / regions.
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=300, linewidth=1, marker='o', edgecolors='black', facecolors='none')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
#Plot data, support vectors, decision boundary and margins
plt.scatter(X[:,0], X[:,1], c=y)
plot_svc_decision_function(model)
```

```python
#Print the support vectors
model.support_vectors_
```

```python
#Only the position of the support vectors matter; 
#Any points further from the margin which are on the correct side do not modify the fit!

X, y = make_blobs(n_samples=200, centers=2, random_state=0, cluster_std=0.6)
model = SVC(kernel='linear', C=1).fit(X,y)

plt.scatter(X[:,0], X[:,1], c=y)
plot_svc_decision_function(model)
```

```python
#Non linearly separable data

from sklearn.datasets import make_circles
X, y = make_circles(100, factor=0.1, noise=0.1, random_state=0)

plt.scatter(X[:,0], X[:,1], c=y);
```

```python
#Try with linear kernel

model = SVC(kernel='linear', C=1).fit(X,y)

plt.scatter(X[:, 0], X[:, 1], c=y)
plot_svc_decision_function(model, plot_support=False)
```

```python
#Apply RBF kernel

model = SVC(kernel='rbf', C=1)
model.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y)
plot_svc_decision_function(model, plot_support=True)
```

```python
#Apply polynomial kernel - Quadratic function

model = SVC(kernel='poly', degree=2, C=1)
model.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y)
plot_svc_decision_function(model, plot_support=True)
```

```python
#Effect of regularization parameter C

X, y = make_blobs(n_samples=100, centers=2, random_state=0)
plt.scatter(X[:, 0], X[:, 1], c=y);
```

```python
#Low C value - larger margin

model = SVC(kernel='linear', C=0.1).fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y)
plot_svc_decision_function(model, plot_support=False)
```

```python
#High C Value

model = SVC(kernel='linear', C=10.0).fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y)
plot_svc_decision_function(model, plot_support=False)
```
